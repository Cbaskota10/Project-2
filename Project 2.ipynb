```python
#Here we will select a random word from a list of words
import random

#List of words for which the user will try to guess
words = ['WHDSPQZ', 'XHO', 'TTDBFRT', 'QQJYF', 'ENQD', 'DNPK', 'CNTR', 'EHWHOD', 'TSVMOHOF', 'XNOCFQGTM']

#This function will choose one random word from the list of words above
word = random.choice(words)
print(encrypted(word))
print("Guess an uppercase letter:")
guesses = ''

#This defines number of turns user will have
turns = 10

while turns > 0: 
    
    #This counts the number of times the user gets the guess incorrect
    failed = 0
    
    #The for loop takes all characters from the user input 
    for char in word: 
        
        #comparing characters with the character in guesses
        if char in guesses:
            print(char)
        else: 
            print ("__")
            
            #for every incorrect guess, failed will be incremented by 1
            failed += 1
        
    if failed ==0: 
        
        #User shall win the game if failure = 0 and will be congragulated upon winning
        print("Hurrah! You won.")
        
        #Prints out the correct word
        print("The correct word is", word)
        break
    
    #if user inputs an incorrect character, it will tell user to input another letter again
    guess = input("Guess an uppercase letter:")
    
    #every user input character will be stored in guesses and will be recorded
    guesses += guess
    
    #checks if user input (letter) is in word or not
    if guess not in word:
        
        #if letter not in word, turns decrease by -1
        turns -= 1
        
        #user will be informed with this message saying it is wrong
        print("Wrong!")
        
        #prints the number of turns the user has left
        print("You have", + turns, "more guesses")
        
        #if user has 10 wrong guesses (turns=0), then they will be prompted with this game over message
        if turns == 0:
            print ("Game Over! The word was", word)

print(decrypt(word))
         
    
```
